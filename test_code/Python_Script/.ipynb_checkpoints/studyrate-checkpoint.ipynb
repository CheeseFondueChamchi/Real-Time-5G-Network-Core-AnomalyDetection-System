```python
import csv
import pymysql
import pandas as pd
import os
from multiprocessing.pool import ThreadPool

# Set the number of threads for multi-threading
num_threads = 4

def GetData(args):
    query, db, table_name = args

    cnx = pymysql.connect(user='traffic', password='traffic1', host='172.21.62.196', database=db)
    cursor = cnx.cursor()
    cursor.execute(query)

    tables = cursor.fetchall()
    cursor.close()
    cnx.close()
    return (db, table_name, tables)

def ProcessTable(args):
    db, table_name = args
    print("DB",db)
    print("table_name",table_name)

    table_name2 = table_name.replace("|", "_")
    csv_filename = f"/home/infra/Documents/NF_table_description/{db}/{table_name2}.csv"

    if os.path.exists(csv_filename):
        return

    query = f"SELECT COLUMN_NAME FROM INFORMATION_SCHEMA.COLUMNS WHERE TABLE_SCHEMA = '{db}' AND TABLE_NAME = '{table_name}'"
    columns = GetData((query, db, table_name))[2]

    if len(columns) == 0:
        return

    columns = [column[0] for column in columns]

    cnx = pymysql.connect(user='traffic', password='traffic1', host='172.21.62.196', database=db)
    cursor = cnx.cursor()
    cursor.execute(f"SELECT * FROM `{table_name}`")
    df = cursor.fetchall()
    cursor.close()    
    cnx.close()
    
    df = pd.DataFrame(df, columns=columns)
    ports = df["PORT"].unique()
    
    table_df = pd.DataFrame()
    for column in columns:
        # Skip the 'PORT' column
        if column in ['DATETIME', 'SYSTEM', 'PORT']:
            continue


        for port in ports:
            tmp_df = df.loc[df["PORT"] == port, column]

            col_min = tmp_df.min()
            col_max = tmp_df.max()

            try:
                col_avg = tmp_df.mean()
                col_var = tmp_df.var()
            except TypeError:
                time_diff_seconds = tmp_df.diff().dt.total_seconds()
                time_diff_minutes = time_diff_seconds / 60
                col_avg = time_diff_minutes.mean()
                col_var = time_diff_minutes.max()

            row_df = pd.DataFrame({'DB_name': db, 'table_name': table_name, 'port': port, 'col_name': column, 'col_min': col_min, 'col_max': col_max, 'col_avg': col_avg, 'col_var': col_var,'shape': [tmp_df.shape]})
            table_df = pd.concat([table_df, row_df], axis=0)

    table_df.to_csv(csv_filename, index=False)
    
if __name__ == '__main__':
db_list = ['BS_UPF01_5M', 'GR_UPF06_5M','BS_AMF21_5M', 'BS_AMF22_5M', 'BS_SMF02_5M', 'BS_SMF03_5M',  'DG_UPF01_5M',
           'DJ_AMF21_5M', 'DJ_AMF22_5M', 'DJ_SMF01_5M', 'DJ_SMF03_5M', 'DJ_UPF01_5M', 'DJ_UPF03_5M', 'GJ_UPF01_5M',
           'GR_AMF21_5M', 'GR_AMF22_5M', 'GR_SMF01_5M', 'GR_SMF04_5M', 'GR_UPF01_5M', 'GR_UPF03_5M', 'GR_UPF05_5M',
           'BS_UPF03_5M', 'HH_AMF21_5M', 'HH_AMF22_5M', 'HH_SMF01_5M', 'HH_SMF04_5M', 'HH_UPF01_5M', 'HH_UPF03_5M',
           'HH_UPF04_5M', 'HH_UPF05_5M', 'HH_UPF06_5M']

# Open a single connection to the database


thread_pool = ThreadPool(num_threads)

for db in db_list:
    cnx = pymysql.connect(user='traffic', password='traffic1', host='172.21.62.196')
    cursor = cnx.cursor()
    query = f"SELECT TABLE_NAME FROM INFORMATION_SCHEMA.TABLES WHERE TABLE_SCHEMA = '{db}'"
    cursor.execute(query)
    tables = cursor.fetchall()
    cursor.close()
    cnx.close()
    
    if not os.path.exists(f"/home/infra/Documents/NF_table_description/{db}"):
        os.makedirs(f"/home/infra/Documents/NF_table_description/{db}")

    args = []

    for table_name in tables:
        table_name = table_name[0]
        table_check_list = ['PUD|PKT', 'CALL|SESS', 'PPORT|PKT', 'DPLD|DPLD', 'PFCP|PFMC', 'SMDNN|PBA',
                            'SMDNN|SE', 'DIAMN|PCRF', 'DIAMN|OCS', 'COSVC|SMPOL', 'COSVC|AMFCOM', 'RAD|ACCT',
                            'PFCP|PFMC', 'PDUSMF|SE', 'PDUSMF|SM', 'PDUSMF|SR', 'REG|IPLMN', 'SR|IPLMN', 'SR|IPLMN',
                            'COSVC|UDMUECM', 'COSVC|UEAUTH', 'COSVC|NRFNFD', 'COSVC|EIR', 'PAG|SM1', 'PAG|SM2']

        if table_name not in table_check_list:
            continue

        args.append((db, table_name))
    print(args)
    thread_pool.map(ProcessTable, args)
```
